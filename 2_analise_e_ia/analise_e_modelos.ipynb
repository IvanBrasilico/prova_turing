```python
import numpy as np
print('Indexing word vectors.')

embeddings_index = {}
with open('glove_s100.txt') as f:
    for line in f:
        values = line.split()
        word = values[0]
        try:
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
        except ValueError:
            pass

print('Found %s word vectors.' % len(embeddings_index))
```

```python
with open('embeddings_glove_pt100.pkl', 'wb') as emb_out:
    pickle.dump(embeddings_index, emb_out)
```
